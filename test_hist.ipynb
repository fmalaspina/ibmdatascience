```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Crear un DataFrame con datos de ejemplo
data = {
    'Edad': np.random.normal(30, 5, 1000),  # Distribución normal con media 30 y desviación estándar 5
    'Ingresos': np.random.normal(50000, 15000, 1000),  # Distribución normal con media 50000
    'Gastos': np.random.normal(20000, 5000, 1000)  # Distribución normal con media 20000
}

df = pd.DataFrame(data)

# Graficar histogramas para todas las columnas
df.hist()
plt.show()
```
